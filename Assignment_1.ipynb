```python
# 8. Write a program to demonstrate the use of logical operators
print("--- 'and' Operator ---")

# The 'and' operator returns True only if both operands are True.
a = True
b = True
print(a and b)

c = True
d = False
print(c and d)

e = False
f = False
print(e and f)

# The 'or' operator returns True if at least one of the operands is True.
print("--- 'or' Operator ---")
a = True
b = True
print(a or b)

c = True
d = False
print(c or d)

e = False
f = False
print(e or f)

# The 'not' operator inverts the Boolean value of the operand
print("--- 'not' Operator ---")
a = True
print(not a)

b = False
print(not b)

```

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types
string_value = "678"
print(string_value)
print(type(string_value))

# changing string to integer
integer_value = int(string_value)
print(integer_value)
print(type(integer_value))

# changing string to float
float_value = float(string_value)
print(float_value)
print(type(float_value))

# changing string to boolean
boolean_value = bool(string_value)
print(boolean_value)
print(type(boolean_value))



```

```python
# 11. Write a program that checks if a number is positive, negative, or zero.
user_input = input("Enter a number: ")
number = float(user_input)

# Use an if-elif-else structure to check the number's value.
if number > 0:
    print(f"The number {number} is positive.")
elif number < 0:
    print(f"The number {number} is negative.")
else:
    print(f"The number {number} is zero.")
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50
total_sum = 0
for number in range(2, 51, 2):
    total_sum += number  # Add the current even number to the total.

# Print the final result.
print(f"The sum of all even numbers between 1 and 50 is: {total_sum}")
```
